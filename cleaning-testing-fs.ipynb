### Handling Categorical Columns with One-Hot Encoding

```python
train_data.dtypes.tolist()
```

```python
one_hot_train_data = pd.get_dummies(train_data, columns=['ethnicity', 'gender', 'hospital_admit_source', 'icu_admit_source', 'icu_stay_type', 'icu_type'])
```

```python
one_hot_train_data.columns.tolist()
```

### Filling out NaN values with the mean of each column

```python
one_hot_train_data.fillna(one_hot_train_data.mean(), inplace=True)
```

### Normalizing the data so all numerical values are between 0 and 1

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        if (max_value - min_value) == 0:
            result[feature_name] = result[feature_name]
        else:
            result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```

```python
cleaned_train_data = normalize(one_hot_train_data)
```

### Feature Selection

Based on [this article](https://machinelearningmastery.com/feature-selection-with-real-and-categorical-data/), feature selection based on correlation statistics are typically based on the nature of the input and output variables (i.e. whether they are categorical or numerical). Since my variables are a mix of both, I decided to select my features by using the feature importance property of the LGBM model. I chose the LGBM model for this purpose because it runs most quickly and gave one of the highest AUC scores from the previous round.

##### Feature Selection with LGBM

```python
cleaned_train_data = cleaned_train_data.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
X, y = cleaned_train_data.drop(columns=['diabetes_mellitus']), cleaned_train_data['diabetes_mellitus']
```

```python
gbm = lgb.LGBMRegressor()
gbm.fit(X, y)
gbm.booster_.feature_importance()
```

```python
fea_imp_ = pd.DataFrame({'cols':X.columns, 'fea_imp':gbm.feature_importances_})
fea_imp_.loc[fea_imp_.fea_imp > 0].sort_values(by=['fea_imp'], ascending = False).head(50)
```
